```python
bar=t.groupby("Channel")["Detergents_Paper"].mean().plot(kind="bar",rot=0,color=["red","Blue"])
value=t.groupby("Channel")["Detergents_Paper"].mean().values
for i,v in enumerate(value):
    mp.text(i,v+0.5,str(v),ha="center",va="bottom")
mp.xlabel("Channel-Wise Mean Detergents_Paper Value");
```

```python
fig,ax=mp.subplots(6,2,figsize=(10,12))
col=["Fresh","Milk","Grocery","Frozen","Detergents_Paper","Delicatessen"]
for i in range(6):
    sn.distplot(t[col[i]],ax=ax[i,0])
    sm.qqplot(t[col[i]],line="s",ax=ax[i,1]) 
    ax[i,0].set_ylabel(col[i],fontsize=14,rotation=0,labelpad=30)
    ax[i,0].set_xlabel("KDE",fontsize=14)
    ax[i,0].set_title("")
    ax[i,0].set_xticks([])
    ax[i,0].set_yticks([])
    ax[i,1].set_xlabel("QQPLOT",fontsize=14)
    ax[i,1].set_ylabel("")
    ax[i,1].set_title("")
    ax[i,1].set_xticks([])
    ax[i,1].set_yticks([])
```

```python
t.groupby(["Channel","Region"])["Milk"].mean().unstack().plot(kind="bar",rot=0,stacked=True)
grouped_data = t.groupby(["Channel","Region"])["Milk"].mean().unstack()
cumulative_sums = grouped_data.cumsum(axis=1)
for i, row in grouped_data.iterrows():
    for col in grouped_data.columns:
        value = row[col]
        if i==1:
            x_pos=i/30
        else:
            x_pos = i/2
        y_pos = (cumulative_sums[col][i] - grouped_data[col][i] / 2)
        mp.text(x_pos, y_pos,f"{value:.2f}", ha='center', va='center')
mp.xlabel("Channel-Region Wise Mean Milk Value");
```
